***Metric Evaluation***

Pada model yang akan dibangun, proses pembersihan data hanya akan menghapus outliers yang bernilai sangat ekstrem, sehingga data tetap memungkinkan memiliki sejumlah outliers yang tidak dihilangkan sepenuhnya. Kondisi ini umum terjadi pada data harga properti karena adanya perbedaan nilai apartemen yang cukup signifikan berdasarkan lokasi, ukuran, dan fasilitas. Oleh karena itu, pemilihan metric evaluation difokuskan pada metrik yang tidak terlalu sensitif terhadap outliers. Metric seperti MSE dan RMSE tidak digunakan sebagai metrik utama karena sangat sensitif terhadap nilai ekstrem dan dapat memberikan penalti yang terlalu besar pada kesalahan prediksi. Adapun metric evaluation yang digunakan dalam proyek ini adalah sebagai berikut.

1. R-Square
<br>
<br>
Metric R-Square digunakan untuk mengukur seberapa besar variasi harga apartemen sebagai target dapat dijelaskan oleh seluruh fitur yang digunakan dalam model. Nilai R-Square merepresentasikan proporsi pengaruh fitur terhadap target, di mana semakin tinggi nilai R-Square maka semakin baik kemampuan model dalam menjelaskan pola harga apartemen. Metric ini bernilai antara 0 hingga 1 dan digunakan terutama pada tahap pemilihan base model untuk membandingkan performa antar model regresi secara umum.

2. MAE (Mean Absolute Error)
<br>
<br>
MAE digunakan untuk mengukur rata-rata selisih absolut antara nilai prediksi harga apartemen dengan harga aktual. Metric ini dipilih karena tidak terlalu sensitif terhadap outliers dan mudah diinterpretasikan dalam konteks bisnis properti. Dengan menggunakan nilai absolut, MAE tidak memperhitungkan arah kesalahan (positif atau negatif), sehingga memberikan gambaran yang jelas mengenai besar kesalahan prediksi rata-rata. Semakin kecil nilai MAE, maka semakin baik performa model dalam memprediksi harga apartemen.

3. MAPE (Mean Absolute Percentage Error)
<br>
<br>
MAPE digunakan untuk mengukur rata-rata kesalahan prediksi dalam bentuk persentase terhadap harga aktual. Metric ini membantu dalam memahami seberapa besar kesalahan prediksi relatif terhadap nilai harga apartemen yang sebenarnya. Berbeda dengan MAE yang bersifat absolut, MAPE mengubah selisih prediksi menjadi bentuk persentase, sehingga lebih mudah dipahami oleh stakeholder non-teknis. Nilai MAPE berada pada rentang 0 hingga 1, di mana nilai yang lebih kecil menunjukkan performa prediksi yang lebih baik.

4. RMSLE (Root Mean Squared Log Error)
<br>
<br>
RMSLE digunakan karena data harga apartemen cenderung memiliki distribusi yang right skewed dengan rentang nilai yang cukup lebar. Metric ini bekerja dengan mengubah nilai harga ke dalam skala logaritmik sebelum menghitung kesalahan, sehingga perbedaan harga yang sangat besar tidak memberikan dampak berlebihan terhadap hasil evaluasi. RMSLE sangat cocok digunakan untuk kasus prediksi harga properti karena lebih menekankan pada kesalahan relatif dibandingkan kesalahan absolut, sehingga memberikan evaluasi performa model yang lebih stabil dan representatif.

##### 3.4.3. *Extreme Outliers*

Keberadaan outlier dapat memengaruhi performa model machine learning. Namun, tidak semua outlier harus dihapus. Pada tahap ini, hanya outlier yang tergolong ekstrem yang akan diidentifikasi menggunakan metode IQR.

cek seluruh data yang memiliki data *outlier* yang hanya memiliki nilai numerik saja.

```python
def check_outlier(col):
    Q1 = np.percentile(col, 25)
    Q3 = np.percentile(col, 75)
    
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return col[(col < lower_bound) | (col > upper_bound)]
```

```python
feature_outlier = []

for col in df.select_dtypes(include=['int64', 'float64']):
    outliers = check_outlier(df[col])
    if len(outliers) > 0:
        print(f"Feature {col} memiliki {len(outliers)} outlier")
        feature_outlier.append(col)
    else:
        print(f"Feature {col} tidak memiliki outlier")
```

Berdasarkan correlation matrix di atas, terlihat bahwa feature `Size(sqf)` dan `YearBuilt` memiliki korelasi positif yang relatif kuat terhadap target `SalePrice`. Hal ini menunjukkan bahwa apartemen dengan ukuran lebih besar dan tahun pembangunan yang lebih baru cenderung memiliki harga yang lebih tinggi. Selanjutnya, dilakukan pengecekan multicollinearity untuk memastikan tidak terdapat hubungan linear yang terlalu kuat antar feature independen.

Multicollinearity diperiksa menggunakan Variance Inflation Factor (VIF). Nilai VIF yang tinggi menunjukkan adanya korelasi kuat antar feature independen, yang berpotensi memengaruhi stabilitas model regresi.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    vif = pd.DataFrame()
    vif['variables'] = X.columns
    vif['VIF'] = [
        variance_inflation_factor(X.values, i)
        for i in range(X.shape[1])
    ]
    return vif
```

```python
X_vif = df[
    ['YearBuilt', 'Size(sqf)', 'N_Parkinglot(Basement)']
]

calc_vif(X_vif)
```

Berdasarkan hasil perhitungan VIF, beberapa feature memiliki nilai VIF di atas batas umum (VIF > 4), yang mengindikasikan adanya korelasi antar feature. Namun, feature seperti `YearBuilt` dan `Size(sqf)` tetap dipertahankan karena memiliki pengaruh yang kuat dan relevan secara domain knowledge terhadap harga apartemen. Oleh karena itu, tidak dilakukan penghapusan feature pada tahap ini.

Proses encoding akan dijalankan bersamaan dengan proses modeling melalui pipeline, sehingga tidak terjadi data leakage antara data training dan data testing.

#### 4.2 *Splitting the Data*

Pada tahap ini dilakukan pemisahan dataset menjadi data latih (training data) dan data uji (testing data). Proses ini bertujuan untuk mengevaluasi performa model pada data yang belum pernah dilihat sebelumnya. Proporsi data yang digunakan adalah 70% untuk data latih dan 30% untuk data uji.

```python
X = df.drop(['SalePrice'], axis=1)
y = df['SalePrice']


X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=100
)
```

```python
print("Jumlah data training:", X_train.shape)
print("Jumlah data testing :", X_test.shape)
```

### 4.3 *Benchmark the Models*

Pada tahap ini dilakukan proses benchmark terhadap beberapa model machine learning untuk menentukan model dasar (base model) terbaik. Proses benchmark dilakukan dalam tiga tahap, yaitu pemilihan base model menggunakan data train, pengujian base model terbaik pada data test, serta perbandingan performa antara data train dan data test berdasarkan metric evaluation.

##### 4.3.1 *Choose Base Model*

Tahap pertama dilakukan dengan melatih seluruh model menggunakan data train dan mengevaluasi performanya menggunakan cross validation. Pada tahap ini digunakan TransformedTargetRegressor untuk melakukan transformasi logaritmik pada target agar distribusi data lebih mendekati normal dan meningkatkan performa model.

```python
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=100)
rf = RandomForestRegressor(random_state=100)
gbt = GradientBoostingRegressor(random_state=100)
xgb = XGBRegressor(random_state=100, verbosity=0)

ttr_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
ttr_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
ttr_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
ttr_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
ttr_gbt = TransformedTargetRegressor(gbt, func=np.log, inverse_func=np.exp)
ttr_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

models = [ttr_lr, ttr_knn, ttr_dt, ttr_rf, ttr_gbt, ttr_xgb]

score_mae = []
score_mape = []
score_rmsle = []
score_r2 = []

kf = KFold(n_splits=5, shuffle=True, random_state=100)

for model in models:
    estimator = Pipeline([
        ('preprocessing', ct),
        ('model', model)
    ])

    mae = cross_val_score(
        estimator, X_train, y_train,
        cv=kf, scoring='neg_mean_absolute_error'
    ).mean()
    score_mae.append(mae)

    mape = cross_val_score(
        estimator, X_train, y_train,
        cv=kf, scoring='neg_mean_absolute_percentage_error'
    ).mean()
    score_mape.append(mape)

    rmsle = cross_val_score(
        estimator, X_train, y_train,
        cv=kf, scoring='neg_mean_squared_log_error'
    ).mean()
    score_rmsle.append(rmsle)

    r2 = cross_val_score(
        estimator, X_train, y_train,
        cv=kf, scoring='r2'
    ).mean()
    score_r2.append(r2)

score_rmsle = np.sqrt(abs(np.array(score_rmsle)))
```

Hasil *metric evaluation* tiap model

```python
model_name = [
    'Linear Regression',
    'KNN Regression',
    'Decision Tree Regression',
    'Random Forest Regression',
    'Gradient Boost Regression',
    'Extreme Gradient Boost'
]

train_score = pd.DataFrame({
    'Model': model_name,
    'Score MAE': score_mae,
    'Score MAPE': score_mape,
    'Score RMSLE': score_rmsle,
    'Score R2': score_r2
})

train_score
```

Berdasarkan hasil benchmark pada data train, Gradient Boost Regression dan Extreme Gradient Boost menunjukkan performa yang sangat berdekatan pada seluruh metric evaluasi. Selisih nilai R-Square di antara kedua model sangat kecil sehingga dapat dianggap tidak signifikan. Oleh karena itu, Extreme Gradient Boost dipilih sebagai base model karena memiliki fleksibilitas yang lebih baik untuk proses hyperparameter tuning serta kemampuan menangani pola non-linear yang lebih kompleks.

Selanjutnya dilakukan pengujian base model terbaik (Extreme Gradient Boost) pada data test untuk melihat performa model pada data yang belum pernah dilihat sebelumnya.

```python
ttr_xgb_test = TransformedTargetRegressor(
    xgb, func=np.log, inverse_func=np.exp
)

models_test = {
    'Extreme Gradient Boost Test': ttr_xgb_test
}

score_mae_test = []
score_mape_test = []
score_rmsle_test = []
score_r2_test = []

for i in models_test:
    model = Pipeline([
        ('preprocessing', ct),
        ('model', models_test[i])
    ])

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    score_mae_test.append(mean_absolute_error(y_test, y_pred))
    score_mape_test.append(mean_absolute_percentage_error(y_test, y_pred))
    score_rmsle_test.append(np.sqrt(mean_squared_log_error(y_test, y_pred)))
    score_r2_test.append(r2_score(y_test, y_pred))

score_before_tuning = pd.DataFrame({
    'Score MAE': score_mae_test,
    'Score MAPE': score_mape_test,
    'Score RMSLE': score_rmsle_test,
    'Score R2': score_r2_test
}, index=models_test.keys())

score_before_tuning
```

Perbandingan Nilai *Train* dan *Test*

```python
hasil_train = train_score.iloc[[5]].set_index('Model')
merged = pd.concat([hasil_train, score_before_tuning])
merged.index = ['Extreme Gradient Boost Train', 'Extreme Gradient Boost Test']
merged
```

Berdasarkan perbandingan nilai metric evaluation pada data train dan data test, terlihat bahwa selisih performa antar keduanya relatif kecil. Hal ini menunjukkan bahwa model tidak mengalami overfitting dan mampu melakukan generalisasi dengan baik. Tahap selanjutnya adalah melakukan hyperparameter tuning untuk meningkatkan performa model lebih lanjut.

##### 4.3.2 *Hyperparameter Tuning*

Pada tahap kedua ini dilakukan proses hyperparameter tuning untuk mendapatkan kombinasi parameter terbaik pada model terpilih, yaitu Extreme Gradient Boost. Model Extreme Gradient Boost memiliki jumlah parameter yang cukup banyak, sehingga tidak seluruh parameter diubah untuk menghindari risiko overfitting.

Pada proses ini, parameter yang dilakukan tuning hanya mencakup:
1. `max_depth` sebagai kedalaman pohon,
2. `learning_rate` sebagai pengatur bobot pembelajaran,
3. `n_estimators` sebagai jumlah iterasi pembelajaran.

```python
max_depth = list(np.arange(7, 11, 1))
learning_rate = [0.01, 0.05, 0.1, 0.5, 0.9]
n_estimators = list(np.arange(75, 301, 25))

hyperparam_space_xgb = {
    'model__max_depth': max_depth,
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators
}
```

```python
xgb = XGBRegressor(
    random_state=100,
    verbosity=0
)

estimator_xgb = Pipeline([
    ('preprocessing', ct),
    ('model', xgb)
])

crossval = KFold(
    n_splits=5,
    shuffle=True,
    random_state=1
)

gs = GridSearchCV(
    estimator=estimator_xgb,
    param_grid=hyperparam_space_xgb,
    cv=crossval,
    scoring=[
        'neg_mean_absolute_error',
        'neg_mean_absolute_percentage_error'
    ],
    refit='neg_mean_absolute_error',
    n_jobs=-1
)
```

```python
gs.fit(X_train, y_train)
```

```python
print('Extreme Gradient Boost')
print('Best_score:', gs.best_score_)
print('Best_params:', gs.best_params_)
```

Berdasarkan hasil hyperparameter tuning, diperoleh kombinasi parameter terbaik sebagai berikut:

1. learning_rate = 0.05  
2. max_depth = 8  
3. n_estimators = 125  

Parameter terbaik tersebut selanjutnya akan digunakan pada model Extreme Gradient Boost untuk dilakukan pengujian ulang pada data test guna melihat peningkatan performa model setelah proses tuning.

##### 4.3.3 *Test Model after Hyperparameter Tuning*

Pada tahap ketiga ini dilakukan pengujian ulang model Extreme Gradient Boost yang telah melalui proses hyperparameter tuning pada data test. Tujuan dari tahap ini adalah untuk mengevaluasi peningkatan performa model setelah dilakukan tuning serta memastikan model mampu melakukan generalisasi dengan baik pada data yang belum pernah dilihat sebelumnya.

```python
xgb_tuning = TransformedTargetRegressor(
    gs.best_estimator_,
    func=np.log,
    inverse_func=np.exp
)

xgb_tuning.fit(X_train, y_train)
y_pred_xgb_tuning = xgb_tuning.predict(X_test)
mae_xgb_tuning = mean_absolute_error(y_test, y_pred_xgb_tuning)
mape_xgb_tuning = mean_absolute_percentage_error(y_test, y_pred_xgb_tuning)
rmsle_xgb_tuning = np.sqrt(mean_squared_log_error(y_test, y_pred_xgb_tuning))

score_after_tuning = pd.DataFrame({
    'Score MAE': mae_xgb_tuning,
    'Score MAPE': mape_xgb_tuning,
    'Score RMSLE': rmsle_xgb_tuning
}, index=['Extreme Gradient Boost Tuning'])

print('Nilai hasil tuning:')
score_after_tuning
```

Berikut adalah perbandingan nilai metric evaluation pada data test antara model sebelum dilakukan hyperparameter tuning dan model setelah dilakukan hyperparameter tuning.

```python
pd.concat([score_before_tuning, score_after_tuning]).drop(columns='Score R2', errors='ignore')
```

Berdasarkan tabel di atas, terlihat bahwa model Extreme Gradient Boost setelah dilakukan hyperparameter tuning menghasilkan nilai MAE, MAPE, dan RMSLE yang lebih rendah dibandingkan sebelum tuning. Hal ini menunjukkan bahwa proses hyperparameter tuning berhasil meningkatkan performa model dalam memprediksi harga apartemen pada data test.

Selanjutnya dilakukan visualisasi perbandingan antara nilai harga aktual pada data test dengan nilai hasil prediksi model setelah dilakukan hyperparameter tuning.

```python
plt.figure(figsize=(14, 10))
sns.regplot(
    x=y_test,
    y=y_pred_xgb_tuning,
    scatter_kws={'color': 'purple', 'alpha': 0.3},
    line_kws={'color': '#CCCC00', 'alpha': 1}
).set(
    title='Actual vs Prediction Price',
    xlabel='Actual Price',
    ylabel='Predicted Price'
)
plt.show()
```

Untuk melihat performa model secara lebih detail, dilakukan analisis metric evaluation berdasarkan rentang harga tertentu pada data test. Analisis ini bertujuan untuk mengetahui bagaimana performa model pada segmen harga rendah hingga tinggi.

```python
df_y_pred = pd.DataFrame({
    'y_test': y_test,
    'y_test_pred': y_pred_xgb_tuning
}).reset_index(drop=True)

ranges = [
    (0, 10000), (10000, 20000), (20000, 30000), (30000, 40000),
    (40000, 50000), (50000, 100000), (100000, 150000),
    (150000, 200000), (200000, 250000), (250000, 300000),
    (300000, 350000), (350000, 400000), (400000, 450000),
    (450000, df_y_pred['y_test'].max())
]

labels = [
    '<10k', '10k-20k', '20k-30k', '30k-40k', '40k-50k',
    '50k-100k', '100k-150k', '150k-200k', '200k-250k',
    '250k-300k', '300k-350k', '350k-400k', '400k-450k', '>450k'
]

list_mae, list_mape, list_len = [], [], []

for (low, high) in ranges:
    temp = df_y_pred[(df_y_pred['y_test'] > low) & (df_y_pred['y_test'] <= high)]
    if len(temp) > 0:
        list_mae.append(mean_absolute_error(temp['y_test'], temp['y_test_pred']))
        list_mape.append(mean_absolute_percentage_error(temp['y_test'], temp['y_test_pred']) * 100)
        list_len.append(len(temp))
    else:
        list_mae.append(0)
        list_mape.append(0)
        list_len.append(0)

range_metrics = pd.DataFrame({
    'Range Price (KRW)': labels,
    'Score MAE (KRW)': list_mae,
    'Score MAPE (%)': list_mape,
    'Jumlah Data': list_len
})

range_metrics
```

Berdasarkan tabel di atas, terlihat bahwa performa model paling optimal berada pada rentang harga menengah, dimana jumlah data relatif lebih banyak. Pada rentang harga yang sangat tinggi, performa model cenderung menurun yang disebabkan oleh keterbatasan jumlah data pada segmen tersebut.

Untuk mengetahui feature apa saja yang paling berpengaruh dalam prediksi harga apartemen, dilakukan analisis feature importance menggunakan model Extreme Gradient Boost setelah tuning.

*Feature Importances*

```python
xgb_final = gs.best_estimator_

feature_imp = pd.DataFrame({
    'Features': ct.get_feature_names_out(),
    'Nilai': xgb_final.named_steps['model'].feature_importances_
}).sort_values('Nilai', ascending=False).head(10)

plt.figure(figsize=(20, 10))
sns.barplot(
    data=feature_imp,
    x='Nilai',
    y='Features',
    color='grey'
)
plt.title('Top 10 Feature Importances')
plt.show()
```

Hasil feature importance menunjukkan bahwa karakteristik fisik apartemen, seperti tipe koridor (HallwayType) dan tahun pembangunan (YearBuilt), merupakan faktor paling berpengaruh dalam penentuan harga apartemen. Hal ini menunjukkan bahwa desain bangunan serta usia apartemen memiliki peranan penting dalam membentuk nilai jual properti.
